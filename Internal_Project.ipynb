# **Global Renewable Energy Consumption Analysis.**

```python
plt.figure
sns.lineplot(data=df, x='Year', y='Energy Consump.')
plt.title('Energy Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.show
```

The line chart shows a decline in renewable energy consumption over the years from 1990 to 2018.

```python
plt.figure
sns.lineplot(data=df, x='Year', y='Energy Consump.', hue='Region')
plt.title('Energy Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.show
```

Renewable energy consumption has declined across all regions, with Sub-Saharan Africa leading as the top consumer, Middle east and North Africa being the lowest consumers. Europe and Central Asia are showing a slight increase in renewable energy adoption.

# **Income Group Analysis**

```python
plt.figure
sns.barplot(data=df, x='Income Group', y='Energy Consump.')
plt.title('Energy Consumption by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Energy Consumption')
plt.show
```


Countries in the low-income group are the largest consumers of renewable energy. Followed closely by lower middle income group.

```python
plt.figure
sns.lineplot(data=df, x='Year', y='Energy Consump.', hue='Income Group')
plt.title('Renewable Energy Consumption by income group')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.show
```

Trend of Energy consumptionby income group over time.
